```python
import os
import glob
import pandas as pd
```

```python
csv_files = glob.glob("one_hidden_mlp_records/*/*trials.csv")
```

```python
def make_dict(file_list):
    data_dict = {}
    for f in file_list:
        # Extract the middle portion: records/<key>/<file>
        key = os.path.basename(os.path.dirname(f))  # folder name
        # Remove '_parzen' or '_random' from the end of the key
        for suffix in ['_parzen', '_random']:
            if key.endswith(suffix):
                key = key[:-len(suffix)]
        data_dict[key] = pd.read_csv(f)
    return data_dict
```

```python
parzen_files = [f for f in csv_files if os.path.basename(os.path.dirname(f)).endswith("parzen")]
random_files = [f for f in csv_files if os.path.basename(os.path.dirname(f)).endswith("random")]
parzen_dict = make_dict(parzen_files)
random_dict = make_dict(random_files)
```
